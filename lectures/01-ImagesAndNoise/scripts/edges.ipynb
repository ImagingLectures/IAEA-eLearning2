```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
```

```python
n=np.linspace(0,30,30)
x,y = np.meshgrid(n-16,n-16)

A = 0.2<=x
fig,ax = plt.subplots(2,2,figsize=(8,8))
ax=ax.ravel()
ax[0].imshow(A, cmap='gray',extent=[0,31,0,31])
ax[0].set_xticks(n[::2])
ax[0].set_xticklabels([]);
ax[0].set_yticks(n[::2])
ax[0].set_yticklabels([]);
ax[0].grid(True,c='r',linewidth=0.5)
ax[0].set_title('Ideal edge')

ax[2].plot(A[0],linewidth=4)
ax[2].axis('off')
ax[2].set_title('Profile of ideal edge');

B = 1/(1+np.exp(-(x)/1.5))
ax[1].imshow((B),cmap='gray',extent=[0,31,0,31])
ax[1].set_title('Smooth edge')
ax[1].set_xticks(n[::2])
ax[1].set_xticklabels([]);
ax[1].set_yticks(n[::2])
ax[1].set_yticklabels([]);
ax[1].grid(True,c='r',linewidth=0.5)

ax[3].plot(B[0],linewidth=4)
ax[3].set_title('Profile of smooth edge');
ax[3].axis('off')

plt.tight_layout()
plt.savefig('edges.svg')
plt.savefig('edges.pdf')
```
