```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
from matplotlib.gridspec import GridSpec

img = np.load('../data/wood.npy')
```

```python
fig=plt.figure(figsize=(12,8))

gs = GridSpec(2, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
ax1.hist(img.ravel(),bins=200)

w1 = [0.5, 0.6]
ax1.plot([w1[0],w1[0],w1[1],w1[1]],[30000,0,30000,0],c='r')
ax2 = fig.add_subplot(gs[1, 0])
ax2.imshow(img,vmin=w1[0],vmax=w1[1],cmap='gray')
ax2.set_title('Interval {0}-{1}'.format(w1[0],w1[1]),c='r')

w2 = [0.8, 0.9]
ax1.plot([w2[0],w2[0],w2[1],w2[1]],[30000,0,30000,0],c='g')
ax3 = fig.add_subplot(gs[1, 1])
ax3.imshow(img,vmin=w2[0],vmax=w2[1],cmap='gray')
ax3.set_title('Interval {0}-{1}'.format(w2[0],w2[1]),c='g')

plt.tight_layout()

plt.savefig('histogram_brightness.svg')
plt.savefig('histogram_brightness.pdf')
```
