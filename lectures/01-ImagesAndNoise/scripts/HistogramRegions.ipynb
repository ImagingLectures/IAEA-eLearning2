```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch

vslice = np.load('../data/wood.npy')
fig,ax=plt.subplots(1,2,figsize=(10,4))

# Do the plotting
ax[0].hist(vslice.ravel(),bins=100)
ax[0].set_xlabel('Pixel value'), ax[0].set_ylabel('Number of pixels')
#a=ax[1].imshow((vslice<0.55).astype(float) - (vslice<0.5).astype(float),interpolation='None');
a=ax[1].imshow(vslice,interpolation='None', cmap = 'gray');
fig.colorbar(a,ax=ax[1],shrink=0.65)

# Patch to the void
con = ConnectionPatch(xyA=(0.98,50000), xyB=(50,50), coordsA="data", coordsB="data", 
                      axesA=ax[0], axesB=ax[1], color="crimson", lw=2)
ax[1].add_artist(con)

con = ConnectionPatch(xyA=(0.8,2500), xyB=(290,480), coordsA="data", coordsB="data", 
                      axesA=ax[0], axesB=ax[1], color="crimson", lw=2)
ax[1].add_artist(con)

con = ConnectionPatch(xyA=(0.52,2500), xyB=(120,270), coordsA="data", coordsB="data", 
                      axesA=ax[0], axesB=ax[1], color="crimson", lw=2)
ax[1].add_artist(con)

#plt.tight_layout()
plt.savefig('histogram_regions.svg')
plt.savefig('histogram_regions.pdf')
```
